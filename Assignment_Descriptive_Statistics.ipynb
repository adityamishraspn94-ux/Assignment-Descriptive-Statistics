### **Q1: Understanding Central Tendency (Easy)**

A bakery tracks its daily muffin sales (in dozens) for one week. The sales data is:
**[10, 12, 11, 15, 14, 13, 12].**

**What is the most representative value of their weekly sales, and why?**

### **Solution:**

The most representative value of the bakery’s weekly muffin sales is the **mean (average)**.

**Mean = (10 + 12 + 11 + 15 + 14 + 13 + 12) / 7
= 87 / 7
≈ 12.43 dozens**
### **Why the mean?**

* The data has **no outliers**.
* The values are **close to each other** and evenly spread.
* In such cases, the **mean best represents the typical weekly sales**.

### **Final Answer:**
**The most representative value is the mean, approximately 12.43 dozens, because the data is evenly distributed without outliers.**

### **Q3: Mode in Real Life (Easy)**

Knowing the mode helps the store manager understand which shoe size is in **highest demand**.

### **Final Answer:**

* This information helps the store manager decide which shoe size to **stock more**, ensuring better inventory planning and reducing out-of-stock situations.

### **Q4: Median in Real Life (Medium)**

A car dealer notes the prices of used cars:
**[$8,000, $9,500, $10,200, $11,000, $50,000].**

**Why is the median a better measure than the mean in this case? Calculate the median.**
### **Solution:**

### **Step 1: Arrange the Data**

Already in order:
**8,000 → 9,500 → 10,200 → 11,000 → 50,000**

### **Step 2: Find the Median**

Median = middle value.
There are 5 values, so the 3rd one is the median.

**Median = $10,200**

### **Step 3: Why is Median Better Than Mean?**

Because the dataset has an **outlier**:

* $50,000 is extremely high compared to the other prices.

This outlier will **pull the mean upward**, making it look like the typical car price is much higher than it really is.

Median is **not affected by extreme values**, so it gives a **more accurate representation** of most car prices.

### **Final Answer:**

**Median = $10,200**

Median is better than mean here because the data has a **very large outlier ($50,000)**, which would distort the mean. The median provides a **more realistic typical price**.

### **Q5: Dispersion Introduction (Medium)**

A student times how long it takes to finish a puzzle each day:
**[25, 30, 27, 35, 40].**

**What does the range tell us about the variation in the student’s puzzle-solving time?**
### **Solution:**

### **Step 1: Calculate the Range**

Range = Highest value − Lowest value
= 40 − 25
= **15 minutes**

### **Step 2: Interpretation**

A range of **15 minutes** means the student's puzzle-solving time **varies quite a bit** from day to day.
Some days the student solves it quickly (25 minutes), and some days it takes much longer (up to 40 minutes).

### **Final Answer:**

The range is **15 minutes**, which shows that the student’s puzzle-solving time has a **moderate to high variation** across the week.

### **Q9: Combining Measures (Hard)**

A family records their monthly electricity usage (in kWh):
**[400, 420, 390, 450, 410].**

**Find the mean and standard deviation. What do these values together tell you about the family’s energy-use pattern?**
### **Solution:**

### **Step 1: Calculate the Mean**

Sum = 400 + 420 + 390 + 450 + 410 = **2070**
Mean = 2070 / 5 = **414 kWh**
### **Step 2: Calculate Standard Deviation (Population SD)**

1. Deviations from mean:

* 400 − 414 = −14
* 420 − 414 = 6
* 390 − 414 = −24
* 450 − 414 = 36
* 410 − 414 = −4

2. Squared deviations:
   196, 36, 576, 1296, 16
   Sum = **2120**

3. Variance = 2120 / 5 = **424**

4. Standard Deviation = √424 ≈ **20.59 kWh**
### **Step 3: Interpretation**

* The **mean usage (414 kWh)** shows the family’s **typical monthly consumption**.
* The **standard deviation (~20.6 kWh)** is small relative to the mean, meaning their electricity usage is **fairly stable** and doesn’t fluctuate much month-to-month
### **Final Answer:**

* **Mean = 414 kWh**
* **Standard deviation ≈ 20.6 kWh**
* Together, these show that the family uses about **414 kWh every month**, with **low variation**, meaning their energy-use pattern is **steady and predictable**.

### **Q10: Practical Application (Hard)**

A basketball player’s points in 8 games are recorded:
**[15, 18, 20, 22, 25, 17, 19, 21].**

**Find the mean, median, mode, range, and standard deviation. What insights can these measures provide about the player’s scoring performance?**
## **Solution:**

### **1. Mean (Average)**

Sum = 157
Mean = 157 / 8 = **19.625 points**

### **2. Median**

Sorted data: **[15, 17, 18, 19, 20, 21, 22, 25]**
Median = (4th + 5th value) / 2
= (19 + 20) / 2
= **19.5 points**

### **3. Mode**

All values appear only once → **No mode**

### **4. Range**

Range = Highest – Lowest = 25 − 15 = **10 points**

### **5. Standard Deviation (Population SD)**

Calculated SD ≈ **2.91 points**

## **Interpretation / Insights**

* **Mean (~19.6)** shows the player typically scores around **20 points per game**.
* **Median (19.5)** is very close to the mean, showing the scoring pattern is **balanced and consistent**.
* **No mode** indicates the player does not repeatedly score the same value; scoring varies naturally.
* **Range (10 points)** shows a moderate difference between the lowest (15) and highest (25) scores.
* **Standard deviation (~2.9)** is relatively small, which means the player’s performance is **stable**—scores don’t fluctuate wildly.
### **Final Answer:**

* **Mean = 19.625**
* **Median = 19.5**
* **Mode = None**
* **Range = 10**
* **Standard Deviation ≈ 2.91**

 Together, these measures show that the basketball player is a **consistent performer**, usually scoring around **20 points per game**, with **low variation** and **steady performance** across matches.
